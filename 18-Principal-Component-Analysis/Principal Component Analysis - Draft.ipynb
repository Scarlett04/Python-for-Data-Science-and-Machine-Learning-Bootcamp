```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df.head()
```

```python
cancer['target']
```

```python
cancer['target_names']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
pd.DataFrame(scaled_data, columns=cancer['feature_names']).head()
```

```python
# PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
print(scaled_data.shape)
print(x_pca.shape)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6), sharey=True)
ax1.scatter(scaled_data[:,0], scaled_data[:,1], c=cancer['target'], cmap='plasma')
ax1.set_xlabel('First Feature')
ax1.set_ylabel('Second Feature')
ax1.set_title('Before PCA')

ax2.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target'], cmap='plasma')
ax2.set_xlabel('First Principal Component')
ax2.set_ylabel('Second Principal Component')
ax2.set_title('After PCA')
```

```python
pca.components_
```

- Rows: Principal component
- Columns: Original feature

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
df_comp
```

```python
sns.heatmap(df_comp, cmap='plasma')
```
