```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips["total_bill"], kde=True, bins=20)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hist')
```

```python
import seaborn as sns
import numpy as np

x = np.random.normal(size=1000)
y = np.random.normal(size=1000)

sns.jointplot(x=x, y=y, kind='hist')
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.violinplot(x='day', y='tip', data=tips, split=False) 
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue="sex", split=True) 
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,
          aspect=1.6, height=10)
```
