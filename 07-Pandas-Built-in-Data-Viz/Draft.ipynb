```python
df1['A'].plot.hist()
```

```python
df2.plot.area(alpha = 0.4)
```

```python
df2.plot.bar(stacked=True)
```

```python
df1.plot.line(y='B',figsize=(12,3),lw=1)
```

```python
# Show by color of column 'C'
df1.plot.scatter(x='A', y='B', c='C')
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
# Show by the size of column 'C'
df1.plot.scatter(x='A', y='B', s=df1['C']*100, cmap='coolwarm')
```

```python
df2.plot.box()
```
